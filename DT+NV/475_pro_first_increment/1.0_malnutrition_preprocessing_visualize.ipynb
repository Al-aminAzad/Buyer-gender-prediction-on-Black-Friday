```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
sns.set_style('whitegrid')
```

```python
df = pd.read_csv('Child_Malnutrition_Dataset.csv')
df = df.drop(['PSU','BORD','HAZ','WAZ','WHZ'],axis=1)
df.head()
```

```python
df.info()
```

```python
df['Division'] = df['Division'].map({1:'barisal',2:'chittagong',3:'dhaka',4:'khulna',5:'Rajshahi',6:'Rangpur',7:'sylhet'})
```

```python
# get division by population
# ---------------------------

plt.figure(figsize=(12,8))
sns.countplot(x='Division',data=df, hue='sex')
plt.ylabel('male and female\n')
plt.title('total male-female in each divison\n')
plt.savefig('total male-female in each divison')

print('most population in chittagong\n------------------------------------------')
print('least population in khulna or barisal\n------------------------------------------')
```

```python
df['Wealth Index'] = df['Wealth Index'].map({1:'poorest',2:'poorer',3:'middle',4:'richer',5:'richest'})
```

```python
# get division by wealth
# ---------------------------

plt.figure(figsize=(12,10))
sns.countplot(x='Division',data=df, hue='Wealth Index')
plt.ylabel('wealth range\n')
plt.title('wealth condition of each divison\n')
plt.savefig('wealth condition of each divison')

print('most richest in dhaka\n---------------------------------------')
print('least richest in barisal/rangpur\n---------------------------------------')
print('most poorest in sylhet\n---------------------------------------')
print('least poorest in khulna\n---------------------------------------')
```

```python
df['wealth_index_cat'] = df['wealth_index_cat'].map({1:'rich',0:'poor'})
```

```python
# get division by wealth
# ---------------------------

plt.figure(figsize=(12,10))
sns.countplot(x='Division',data=df, hue='wealth_index_cat')
plt.ylabel('wealth range\n')
plt.title('wealth condition of each divison\n')
plt.savefig('wealth condition of each divison')

print('most rich in dhaka\n---------------------------------------')
print('least rich in rangpur\n---------------------------------------')
print('most poor in sylhet\n---------------------------------------')
print('least poor in khulna\n---------------------------------------')
```

```python
df.head()
```

```python
# get population by wealth
# -------------------------

df['sex'] = df['sex'].map({1:'female',0:'male'})

plt.figure(figsize=(12,10))
sns.countplot(x='sex',data=df,hue='wealth_index_cat')
plt.title('who is rich')
plt.xlabel('sex')
plt.ylabel('wealth_index_cat')
plt.savefig('who is rich')
```
